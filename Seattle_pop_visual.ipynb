```python
#############################################################################################  
#### Palani's Visualization code Start
#############################################################################################  

%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Age Group classification
Age_group_bins=[0, 4, 7, 9, 13, 17, 18]
Age_group_Names=["0-19 Years", "20-34 Years", "35-44 Years", "45-64 Years", "65-84 Years", "85 and older"]

####
df_plt_us_combine = pd.read_csv("./Data/df_plt_us_combine.csv")

AgeGroup_types = df_plt_us_combine.groupby("AgeGroup_Names")["AgeGroup_Names"].max()
years = df_plt_us_combine.groupby("Year")["Year"].max()


# plotting a graph
plt.figure(figsize=(10,5))

for type in AgeGroup_types:
    if (type=="0-19 Years"):
        color_type = "blue"
    elif (type=="20-34 Years"):
        color_type = "green"
    elif (type=="35-44 Years"):
        color_type = "orange"        
    elif (type=="45-64 Years"):
        color_type = "Yellow"
    elif (type=="65-84 Years"):
        color_type = "purple"        
    else:
        color_type = "red"
    
    plt.plot("Year", "net_change", data=df_plt_us_combine.loc[df_plt_us_combine["AgeGroup_Names"]==type, :], c=color_type, marker="o",   linewidth=2)
 
#plt.ylim(10e7,100e7)
plt.xlim(2000, 2017, 1)
plt.xticks(range(2000, 2019, 1))
#plt.yticks(range(5000000, 100000000, 10000000))
plt.yticks(np.arange(-3, 7.0, 0.5))

# Create a legend
lg_0_19_Years = plt.scatter(1,1, marker="o", color="blue" )
lg_20_34_Years = plt.scatter(1,1, marker="o", color="green" )
lg_35_44_Years = plt.scatter(1,1, marker="o", color="orange" )
lg_45_64_Years= plt.scatter(1,1, marker="o", color="Yellow" )
lg_65_84_Years = plt.scatter(1,1, marker="o", color="purple" )
lg_85_Years_and_older= plt.scatter(1,1, marker="o", color="red" )

plt.grid()

plt.title("Us National Level Net Change in Population By AgeGroup")
plt.xlabel("Years")
plt.ylabel("Net Popluation change (%) by Year by Year ")

plt.legend( (lg_0_19_Years, lg_20_34_Years, lg_35_44_Years, lg_45_64_Years, lg_65_84_Years, lg_85_Years_and_older ), 
("0-19 Years", "20-34 Years", "35-44 Years", "45-64 Years", "65-84 Years", "85 and older"  ), title="AgeGroup Types" )

plt.savefig("./Images/USNational_by_AgeGroup.png")
#plt.show()

############################################################################################# 

df_plt_seattle_combine = pd.read_csv("./Data/df_plt_seattle_combine.csv")

# Plotting the graph
AgeGroup_types = df_plt_seattle_combine.groupby("AgeGroup_Names")["AgeGroup_Names"].max()
years = df_plt_seattle_combine.groupby("Year")["Year"].max()                                     

plt.figure(figsize=(10,5))

for type in AgeGroup_types:
    if (type=="0-19 Years"):
        color_type = "blue"
    elif (type=="20-34 Years"):
        color_type = "green"
    elif (type=="35-44 Years"):
        color_type = "orange"        
    elif (type=="45-64 Years"):
        color_type = "Yellow"
    elif (type=="65-84 Years"):
        color_type = "purple"        
    else:
        color_type = "red"
    
    plt.plot("Year", "net_change", data=df_plt_seattle_combine.loc[df_plt_seattle_combine["AgeGroup_Names"]==type, :], c=color_type, marker="o", linestyle='dashed',  linewidth=2)
    
#plt.ylim(10e7,100e7)
plt.xlim(2000, 2017, 1)
plt.xticks(range(2001, 2019, 1))
#plt.yticks(range(5000000, 100000000, 10000000))
plt.yticks(np.arange(-3, 7.0, 0.5))

# Create a legend
lg_0_19_Years = plt.scatter(1,1, marker="o", color="blue" )
lg_20_34_Years = plt.scatter(1,1, marker="o", color="green" )
lg_35_44_Years = plt.scatter(1,1, marker="o", color="orange" )
lg_45_64_Years= plt.scatter(1,1, marker="o", color="Yellow" )
lg_65_84_Years = plt.scatter(1,1, marker="o", color="purple" )
lg_85_Years_and_older= plt.scatter(1,1, marker="o", color="red" )

plt.grid()

plt.title(f"Seattle Metro Level Net Change in Population By AgeGroup")
plt.xlabel("Years")
plt.ylabel("Net Popluation change (%) by Year by Year ")

plt.legend( (lg_0_19_Years, lg_20_34_Years, lg_35_44_Years, lg_45_64_Years, lg_65_84_Years, lg_85_Years_and_older ), 
("0-19 Years", "20-34 Years", "35-44 Years", "45-64 Years", "65-84 Years", "85 and older"  ), title="AgeGroup Types" )

plt.savefig("./Images/Seattle_Metro_by_AgeGroup.png")

############################################################################################# 
### Population Net change between 2001 and 2019

for type in AgeGroup_types:
    if (type=="0-19 Years"):
        color_type = "blue"        
    elif (type=="20-34 Years"):
        color_type = "green"
    elif (type=="35-44 Years"):
        color_type = "orange"        
    elif (type=="45-64 Years"):
        color_type = "Yellow"
    elif (type=="65-84 Years"):
        color_type = "purple"        
    else:
        color_type = "red"
   
    #fig = plt.figure()  
   
    plot_years = [2001, 2009, 2017]
    #print(df_plt_seattle_combine)
    
    df_seattle_net_changes_by_year = df_plt_seattle_combine.loc[(df_plt_seattle_combine["AgeGroup_Names"]==type) & (df_plt_seattle_combine["Year"].isin(plot_years)),
                                                                ["net_change", "Year", "Sum", "AgeGroup_Names"]]
    #print(df_seattle_net_changes_by_year)
    net_pop_change=[]
    net_change = 0
    prev_sum = 0 
    for year in plot_years:
        ##print(f"year: {year}")
        sum = list(df_seattle_net_changes_by_year.loc[(df_seattle_net_changes_by_year["AgeGroup_Names"] == type) & 
                                                    (df_seattle_net_changes_by_year["Year"]==year), "Sum"])[0]
        ##print(f"Sum: {sum}")
        
        if year == 2001:
            #net_change = list(df_seattle_net_changes_by_year["net_change"])[0]
            net_change=0
            prev_sum = list(df_seattle_net_changes_by_year["Sum"])[0]
            #print(f" {agegrp} : {year}: Sum:{sum} : Prev_sum{prev_sum} ")
        else:
            #print(f" {agegrp} : {year}: Sum:{sum} : Prev_sum{prev_sum} ")
            net_change = ((sum - prev_sum)/prev_sum) * 100
            ##print(f"Net_change-{year}:{net_change}")
            prev_sum = sum
            
        net_pop_change.append(round(net_change,4))
     
    ###print(f"Netpopchange: {net_pop_change}")
    ###print(type)
    df_seattle_net_changes_by_year["net_change"]=net_pop_change
                                  
    ###print(df_seattle_net_changes_by_year)                               
    ###################### 
    df_us_net_changes_by_year = df_plt_us_combine.loc[(df_plt_seattle_combine["AgeGroup_Names"]==type) & (df_plt_us_combine["Year"].isin(plot_years)),
                                                                ["net_change", "Year", "Sum", "AgeGroup_Names"]]
    #print(df_seattle_net_changes_by_year)
    net_pop_change=[]
    net_change = 0
    prev_sum = 0 
    for year in plot_years:
        ##print(f"year: {year}")
        sum = list(df_us_net_changes_by_year.loc[(df_us_net_changes_by_year["AgeGroup_Names"] == type) & 
                                                    (df_us_net_changes_by_year["Year"]==year), "Sum"])[0]
        ##print(f"Sum: {sum}")
        
        if year == 2001:
            #net_change = list(df_us_net_changes_by_year["net_change"])[0]
            net_change=0
            prev_sum = list(df_us_net_changes_by_year["Sum"])[0]
            #print(f" {agegrp} : {year}: Sum:{sum} : Prev_sum{prev_sum} ")
        else:
            #print(f" {agegrp} : {year}: Sum:{sum} : Prev_sum{prev_sum} ")
            net_change = ((sum - prev_sum)/prev_sum) * 100
            ##print(f"Net_change-{year}:{net_change}")
            prev_sum = sum
            
        net_pop_change.append(round(net_change,4))
     
    ###print(f"Netpopchange: {net_pop_change}")
    ###print(type)
    df_us_net_changes_by_year["net_change"]=net_pop_change
    
    ###print(df_us_net_changes_by_year)
    
#################
    df_seattle_us_combine = pd.merge(df_seattle_net_changes_by_year, df_us_net_changes_by_year, how="inner", on="Year").rename(columns = {"net_change_x":"net_change_seattle", "net_change_y":"net_change_us"})
    df_plt_seattle_us_combine = df_seattle_us_combine.loc[df_seattle_us_combine["Year"] > 2001, :].set_index("Year") 
    #df_seattle_us_combine = df_seattle_us_combine.set_index("Year")   
    ###print(df_seattle_us_combine)
    ######df_plt_seattle_us_combine.to_csv(f"./df_plt_seattle_us_combine_{type}.csv")


    df_plt_seattle_us_combine[[ "net_change_seattle", "net_change_us"]].plot(kind="bar",  figsize=(10, 6), color=("b", "orange") )
    
    #plt.xlim(xmin=2001)
    plt.title(f"Seattle Metro Vs US National Net Change in Population By AgeGroup ({type})")
    plt.xlabel("Years")
    plt.ylabel("Net Popluation change (%) by from Year - 2001")
    plt.grid()

    plt.savefig(f"./Images/US_Seattle_Metro_by_AgeGroup_2009_2017_{type}.png")
    
########################################################################################################

df_plt_seattle_combine_gender = pd.read_csv("./Data/df_plt_seattle_combine_gender.csv")
df_plt_us_combine_gender = pd.read_csv("./Data/df_plt_us_combine_gender.csv")

gender_types = df_plt_us_combine_gender.groupby("SEX")["SEX"].max()
years = df_plt_us_combine_gender.groupby("Year")["Year"].max()

for gender in gender_types:
    if (gender==1):
        color_type = "blue"
        gender_name = "Male"
    elif (gender==2):
        color_type = "deeppink"
        gender_name = "Female"
    else:
        color_type = "red"
   
    plt.figure(figsize=(10,5))  
			 
    line1, = plt.plot("Year", "net_change", data=df_plt_seattle_combine_gender.tail(100).loc[df_plt_seattle_combine_gender.tail(100)["SEX"]==gender, :],
             c=color_type, marker="o", linestyle="dashed", linewidth=2, label=f"Seattle {gender_name}")
			
    line2, =plt.plot("Year", "net_change", data=df_plt_us_combine_gender.tail(100).loc[df_plt_us_combine_gender.tail(100)["SEX"]==gender, :],
             c=color_type, marker="o", linewidth=2, label=f"US {gender_name}")
    
    plt.legend(handles=[line1, line2], markerscale=0.1)
    
    plt.title(f"Seattle Metro Vs US National Net Change in Population By Gender ({gender_name})")
    plt.xlabel("Years")
    plt.ylabel("Net Popluation change (%) by Year by Year ")
    plt.grid()

    plt.xlim(2000, 2017, 1)
    plt.xticks(range(2000, 2019, 1))
    #plt.yticks(range(5000000, 100000000, 10000000))
    plt.yticks(np.arange(-0.125, 2.5, 0.125))
    
    plt.savefig(f"./Images/US_Seattle_Metro_by_Gender_{gender_name}.png")
    
#############################################################################################  
#### Palani's Visualization code End
#############################################################################################      
```
